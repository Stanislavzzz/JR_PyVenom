РЎРѕР·РґР°С‚СЊ РЅРѕРІС‹Р№ РїСЂРёР·РЅР°Рє, РєРѕС‚РѕСЂС‹Р№ Р±СѓРґРµС‚ СЃРѕРґРµСЂР¶Р°С‚СЊ С‚РѕР»СЊРєРѕ РєРѕР»-РІРѕ Gb РїР°РјСЏС‚Рё

```python
def convert_memory(x):
    memory = x.split()[0]
    if memory.endswith('GB'):
        result = float(memory.replace('GB', ''))
    elif memory.endswith('TB'):
        result = float(memory.replace('TB', '')) * 1024
    return result
```

```python
df['Memory_GB'] = df['Memory'].apply(convert_memory)
```
